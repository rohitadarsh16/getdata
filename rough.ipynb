```python
async def get_data(pages, concurrent_tasks):
    tasks = set()
    results = []

    async with ClientSession() as session:
        for page in range(1, pages + 1):
            if len(tasks) >= concurrent_tasks:
                results.append(await asyncio.wait(tasks, return_when=asyncio.FIRST_COMPLETED))
                tasks.difference_update({task for task in tasks if task.done()})
            tasks.add(asyncio.ensure_future(get_page_data(session, page)))

        if tasks:
            results.append(await asyncio.gather(*tasks))

    flattened_results = [item for sublist in results for item in sublist]
    return flattened_results

# Run the get_data function with 100 concurrent tasks
data = asyncio.run(get_data(12978, 100))
df = pd.DataFrame(data)
df.to_excel('productdata.xlsx', index=False)
```
